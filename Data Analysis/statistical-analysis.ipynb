````python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

from scipy.stats import spearmanr
````

````python
df = pd.read_csv('/content/drive/MyDrive/GUVI FINAL PROJRCT/RFS_Cleaned_Dataset.csv')
````

### **Unique Values of Features**

````python
def unique_values(columns):
  print(f'Total Rows Count : {df.shape[0]}')
  print()
  for i in columns:
    print(f'{i} : {df[i].nunique()}')
    print()

unique_values(df.select_dtypes(include='number'))
````

## **Pearson Correlation Coefficient (r)**



```
# Here we will do statistic measure  to find strength and direction  of the linear relationship b.w two continous  variable

Range : ( -1 to 1 )

      # value close to 1 means positive correlation if x increase y also increases and vice versa

      # value close to -1 means negative correlation if x increase  y  will decrease and vice versa

      # value close to 0  means no linear relationship
```



````python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(),annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation')
plt.show()
````

## **Spearman Rank Correlation Coeeficient(rho)**

````python
correlation_matrix, _ = spearmanr(df)
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', xticklabels=df.columns, yticklabels=df.columns)
plt.title('Spearman\'s Rank Correlation Heatmap')
plt.show()
````
